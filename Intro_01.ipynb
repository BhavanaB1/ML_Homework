Q7. Sum of weights

Select all the cars from Asia

Select only columns vehicle_weight and model_year

Select the first 7 values

Get the underlying NumPy array. Let's call it X.

Compute matrix-matrix multiplication between the transpose of X and X.

To get the transpose, use X.T. Let's call the result XTX.

Invert XTX.

Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].

Multiply the inverse of XTX with the transpose of X, and then multiple the result by y. Call the result w.

What's the sum of all the elements of the result?

Note: You just implemented linear regression. We'll talk about it in the next lesson.

```python
asia_cars = df[df['origin']=='Asia']
asia_cars
```

```python
subset = asia_cars[['vehicle_weight', 'model_year']]
subset
```

```python
subset = subset.head(7)
subset
```

```python
import numpy as np
```

```python
X = subset.to_numpy()
X
```

```python
XTX = X.T @ X
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y
```

```python
w = XTX_inv @ X.T @ y
w
```

```python
result = w.sum()
result
```
